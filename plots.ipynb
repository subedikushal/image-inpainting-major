```python
import json
import matplotlib.pyplot as plt
```

```python
 
with open('trainlosses.json') as json_file:
    train_losses = json.load(json_file)
 
with open('vallosses.json') as json_file:
    val_losses = json.load(json_file)
```

```python
X_train = list(train_losses.keys())
Y_train = list(train_losses.values())
X_val = list(val_losses.keys())
Y_val = list(val_losses.values())

X_train = list(range(1,11))
X_val = list(range(1,11))
Y_train = [0.024, 0.0091, 0.0080, 0.0069, 0.0060, 0.0057, 0.0042, 0.0032, 0.0024, 0.0016]
Y_val = [0.022, 0.0088, 0.0078, 0.0068, 0.0064, 0.0059, 0.0040, 0.0035, 0.0027, 0.0018]
```

```python
plt.plot(X_train, Y_train)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.legend()
plt.tight_layout()
plt.savefig("training_loss.png")
plt.show()
```

```python
plt.plot(X_val, Y_val)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Validation Loss")
plt.tight_layout()
plt.savefig("validation_loss.png")
plt.show()
```

```python
plt.plot(X_train, Y_train, label="training loss")
plt.plot(X_val, Y_val, label="validation loss")
plt.title("Training Loss vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig("training_vs_validation.png")
plt.show()
```
